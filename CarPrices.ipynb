```python
def sort_exterior_color(row):
  row = str(row).lower()
  if 'crystal' in row or 'diamond' in row:
    return 'crystal'
  elif 'metallic' in row:
    return 'metallic'
  elif 'gray' in row or 'grey' in row or 'nickel' in row:
    return 'gray'
  elif 'silver' in row:
    return 'silver'
  elif 'white' in row or 'pearl' in row or 'chalk' in row or 'powder' in row or 'smoke' in row or 'frostbite' in row or 'billet clearcoat' in row or 'ice' in row:
    return 'white'
  elif 'black' in row or 'charcoal' in row or 'graphite' in row or 'nightfall' in row or 'metal' in row or 'dark' in row or 'caviar' in row or 'platinum' in row or 'magnetic' in row or 'iridium' in row  or 'clearcoat' in row:
    return 'black'
  elif 'blue' in row or 'sea' in row:
    return 'blue'
  elif 'red' in row or 'maroon' in row or 'scarlet' in row or 'fire' in row or 'burgundy' in row:
    return 'red'
  elif 'green' in row or 'alien' in row:
    return 'green'
  elif 'yellow' in row or 'gold' in row or 'mango' in row:
    return 'yellow'
  elif 'brown' in row or 'orange' in row or 'tan' in row or 'mocha' in row:
    return 'orange'
  else:
    return 'others'
df['Exterior color'].apply(sort_exterior_color).value_counts()
```

### Create the engine displacement or no_of_total_litres column

```python
def is_valid_float(input_str):
    try:
        float_value = float(input_str)
        return True
    except ValueError:
        return False


def get_litres(row):
  litre =  row[:3]
  if is_valid_float(litre):
    return litre
  return 0.0

df_clean['no_of_total_Litres'] = df_clean['Engine'].apply(get_litres)
```
